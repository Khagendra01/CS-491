```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np


database = []

for x in range(1, 42):
    src_path = '../DataBook/assignment3_submissions/'
    file_name = f'assignment3-{x}.submit'
    print(file_name)
    src_path += file_name

    with open(src_path, 'r') as assignmentFile:
        json_data = json.load(assignmentFile)
        log_field = json_data['submission']['logs'][0]['log']

    # Split the log_field into individual JSON objects
    logs = log_field.strip().split('\n')

    # Create an empty list to hold dictionaries for each log entry
    log_list = []

    # Iterate over each log entry and append the data to the list
    for log in logs:
        data = json.loads(log)
        log_list.append(data)

    df = pd.DataFrame(log_list)

    time_chars_dict = []
    cnt = 0
    # Processing the DataFrame skipping the first entry
    for timestamp, total_chars, cursor_position in zip(df['t'][1:], df['_cs'].fillna(method='ffill')[1:], df['_c'].fillna(method='ffill')[1:]):
       if not pd.isnull(total_chars) and not pd.isnull(cursor_position):
            time_chars_dict.append( cnt )
            time_chars_dict.append( int(total_chars) )
            time_chars_dict.append( int(cursor_position) )
            cnt += 1

    database.append(time_chars_dict)

#creating a file
output_file = "../DataSet/assignment3New.json"
with open(output_file, 'w') as f:
    json.dump(database, f)

print(f"json saved to {output_file}")

    
# import pandas as pd
# import json
# import matplotlib.pyplot as plt
# import numpy as np


# database = []

# for x in range(1, 55):
#     src_path = '../DataBook/assignment1_submissions/'
#     file_name = f'assignment1-{x}.submit'
#     src_path += file_name

#     with open(src_path, 'r') as assignmentFile:
#         json_data = json.load(assignmentFile)
#         log_field = json_data['submission']['logs'][0]['log']

#     # Split the log_field into individual JSON objects
#     logs = log_field.strip().split('\n')

#     # Create an empty list to hold dictionaries for each log entry
#     log_list = []

#     # Iterate over each log entry and append the data to the list
#     for log in logs:
#         data = json.loads(log)
#         log_list.append(data)

#     df = pd.DataFrame(log_list)

#     time_chars_dict_global = []

#     # Processing the DataFrame skipping the first entry
#     for timestamp, total_chars, cursor_position in zip(df['t'][1:], df['_cs'].fillna(method='ffill')[1:], df['_c'].fillna(method='ffill')[1:]):
#         time_chars_dict = []
#         time_chars_dict.append(timestamp)
#         time_chars_dict.append(total_chars)
#         time_chars_dict.append(cursor_position)
#         time_chars_dict_global.append(time_chars_dict)

#     database.append(time_chars_dict_global)

# #creating a file
# output_file = "../DataSet/assignment1New.json"
# with open(output_file, 'w') as f:
#     json.dump(database, f)

# print(f"Dictionary saved to {output_file}")

    
















```

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np


database = []

for x in range(1, 55):
    src_path = '../DataBook/assignment1_submissions/'
    file_name = f'assignment1-{x}.submit'
    print(file_name)
    src_path += file_name

    with open(src_path, 'r') as assignmentFile:
        json_data = json.load(assignmentFile)
        log_field = json_data['submission']['logs'][0]['log']

    # Split the log_field into individual JSON objects
    logs = log_field.strip().split('\n')

    # Create an empty list to hold dictionaries for each log entry
    log_list = []

    # Iterate over each log entry and append the data to the list
    for log in logs:
        data = json.loads(log)
        log_list.append(data)

    df = pd.DataFrame(log_list)

    char_arr = [[] for _ in range(int(max(df['u'])) + 1)]

    if '^' in df.columns:
        for char, cellNumber in zip(df['^'], df['u']):
            if isinstance(char, list):
                cellNumber = int(cellNumber)
                if len(char) >= 2:
                    char_arr[cellNumber].append(char[2])
    else:
        for char, cellNumber in zip(df['v'], df['u'].fillna(method='ffill')):
            cellNumber = int(cellNumber)
            if isinstance(char, str):
                char_arr[cellNumber].append(char)

    #char_arr[cellNumber].insert(char[0],char[2])
    flattened_arr = [item for sublist in char_arr for item in sublist]
    
    database.append(flattened_arr)

#creating a file
output_file = "../DataSet/text_assignment1.json"
with open(output_file, 'w') as f:
    json.dump(database, f)

print(f"json saved to {output_file}")
           
```
